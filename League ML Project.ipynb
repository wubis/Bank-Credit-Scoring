```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
```

```python
dataChall = pd.read_csv("archive/Challenger_Ranked_Games.csv")
dataGM = pd.read_csv("archive/GrandMaster_Ranked_Games.csv")
dataM = pd.read_csv("archive/Master_Ranked_Games.csv")
```

```python
data = pd.concat([dataChall, dataGM, dataM], axis=0)
data
```

```python
results = data["blueWins"]
results
```

```python
data.drop(columns=["blueWins", "redWins", "redFirstTower", "redFirstBaron", "redFirstDragon", "redFirstInhibitor", "redFirstBlood"], inplace = True)
data
```

```python
data["dBaron"] = data["blueBaronKills"] - data["redBaronKills"]
data["dDragon"] = data["blueDragonKills"] - data["redDragonKills"]
data["dInhib"] = data["blueInhibitorKills"] - data["redInhibitorKills"]
data["dTower"] = data["blueTowerKills"] - data["redTowerKills"]
data["dWardPlace"] = data["blueWardPlaced"] - data["redWardPlaced"]
data["dWardKills"] = data["blueWardkills"] - data["redWardkills"]
data["dKills"] = data["blueKills"] - data["redKills"]
data["dKillingSpree"] = data["blueKillingSpree"] - data["redKillingSpree"]
data["dDamage"] = data["blueChampionDamageDealt"] - data["redChampionDamageDealt"]
data["dGold"] = data["blueTotalGold"] - data["redTotalGold"]
data["dCS"] = data["blueTotalMinionKills"] - data["redTotalMinionKills"]
data["dAvgLevel"] = data["blueAvgLevel"] - data["redAvgLevel"]
data["dJungleCS"] = data["blueJungleMinionKills"] - data["redJungleMinionKills"]
data["dHealing"] = data["blueTotalHeal"] - data["redTotalHeal"]

data.drop(columns=["blueBaronKills", "redBaronKills", "blueDragonKills", "redDragonKills", "blueInhibitorKills", "redInhibitorKills", "blueTowerKills",
                 "redTowerKills", "blueWardPlaced", "redWardPlaced", "blueWardkills", "redWardkills", "blueKills", "redKills", "blueKillingSpree",
                 "redKillingSpree", "blueChampionDamageDealt", "redChampionDamageDealt", "blueTotalGold", "redTotalGold", "blueTotalMinionKills",
                 "redTotalMinionKills", "blueAvgLevel", "redAvgLevel", "blueJungleMinionKills", "redJungleMinionKills", "blueTotalHeal", "redTotalHeal",
                  "blueDeath", "redDeath", "blueAssist", "redAssist", "blueTotalLevel", "redTotalLevel", "redObjectDamageDealt", "blueObjectDamageDealt"], inplace = True)
data
```
